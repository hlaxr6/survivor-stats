```python
from  bs4 import  BeautifulSoup 
import requests
import re
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_columns', 500)
```

```python
def turn_to_url(season):
    return 'http://survivor.wikia.com/wiki/Survivor:_' + season.replace(' ','_')

def process(string):
    stripped = string.strip()
    return re.sub('[0-9]+', '', stripped)

#make sure they are in order
seasons = ['Borneo', 'The Australian Outback', 'Africa', 'Marquesas', 'Thailand', 'The Amazon', 
           'Pearl Islands', 'All-Stars', 'Vanuatu', 'Palau', 'Guatemala', 'Panama', 
           'Cook Islands', 'Fiji', 'China', 'Micronesia', 'Gabon', 'Tocantins', 'Samoa', 
           'Heroes vs. Villains', 'Nicaragua', 'Redemption Island', 'South Pacific',
           'One World', 'Philippines', 'Caramoan', 'Blood vs. Water', 'Cagayan', 'San Juan del Sur',
           'Worlds Apart', 'Cambodia', 'Kaôh Rōng', 'Millennials vs. Gen X',
           'Game Changers', 'Heroes vs. Healers vs. Hustlers', 'Ghost Island','David vs. Goliath']

#seasons = ['Cambodia','Millennials vs. Gen X',
#           'Game Changers',]
```

```python
def scrape_survivor_wiki_season_info():
    data = []
    checklabel = 0
    snum = 0
    
    def add_0(row):
        return str(row['Season No.:']).zfill(2)
       
    for season in seasons:
        information = []
        labels = []
        snum+=1
        print(snum)
        #if snum == 27:
        card = requests.get(turn_to_url(season))
        soup = BeautifulSoup(card.content, 'lxml')
        for x in range(1,11):
            line = soup.select('div.pi-item.pi-data.pi-item-spacing.pi-border-color')[x]


            #missing episode count
            if snum == 33:
                if x == 5:
                    labels.append('No. of Episodes:')
                    information.append(14)
                if x > 4: x+=1
                    
            if snum == 34:
                if x == 5:
                    labels.append('No. of Episodes:')
                    information.append(13)
                if x > 4: x+=1

            #print(x line.select('div')[0].text)
            
            if x < 9:
                label = line.select('h3')[0].text
                info = line.select('div')[0].text
            if x == 10:
                label = line.select('h3')[0].text
                #print(label)
                info = ''
                #print(len(line.select('div')[0].select('a')))
                for y in range(len(line.select('div')[0].select('a'))):
                    info += line.select('div')[0].select('a')[y].text
                    if y < (len(line.select('div')[0].select('a'))-1):
                        info += ', '
            
            if x != 9 and x != 11:
                #print(info,label)
                information.append(info)
                labels.append(label)
        #print(checklabel)
        if checklabel == 0:
            data.append(labels)
            checklabel = 1
        #print(labels)
        data.append(information)
            #row_i = summary_table.select('tr')[i+skip]
            #name = row_i.select('td b a')[0].text  
            #print(name)
            #ageloc = row_i.select('td > small')[0].text
            #daysplacing = row_i.select('td')[lastdat].text
            #data.append({'season': season, 'name': name, 'ageloc': ageloc, 'daysplacing': daysplacing, 'season_number':snum,'players_in_season':ncontest})
    #print(data)
    rr = pd.DataFrame(data)
    rr.columns = rr.iloc[0]
    rr['Season No.:'] = rr.apply(add_0,axis=1)
    rr.drop([0],inplace=True)
    rr.set_index('Season No.:',inplace=True)
    rr.sort_index(inplace=True)
    
    rr.to_csv('/home/analytics/Survivor/seasons.csv', index=True)
    
    return rr


scrape_survivor_wiki_season_info()
```
